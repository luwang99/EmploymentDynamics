```python
df1=df.dropna(subset=['Location'])
```

```python
df1.index = np.arange(0, len(df1))
df1.head()
```

```python
for i in range(len(df1)):
    location=df1.iloc[i,23]
    for k in range(len(bay_area_cities)):
        if bay_area_cities[k].lower() in location.lower():
            #print('yes')
            df1.at[i, 'Location'] = bay_area_cities[k]
            break
        elif k==len(bay_area_cities)-1:
            df1.at[i, 'temp'] = 0
```
